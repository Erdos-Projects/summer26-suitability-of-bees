```python
import pandas as pd
import plotly.express as px
# NOTE: DATE CURRENTLY ONLY FOR ACAMPTOPAPPUS
# This will be updated to all host plants (hopefully) on Wednesday or Thursday this week

# Load data
acamptopappus_symbio_data = pd.read_csv("Plants/Acamptopappus/SymbioData/occurrences.csv")
```

```python
# Recording only columns related to dates and geographic data
acamptopappus_symbio_data_filtered = acamptopappus_symbio_data[["genus","locality","eventDate","decimalLatitude","decimalLongitude"]].copy()

# Convert eventDate to datetime
acamptopappus_symbio_data_filtered["eventDate"] = pd.to_datetime(acamptopappus_symbio_data_filtered["eventDate"], format="mixed")

# Western U.S. bounding box
lat_min = 25
lat_max = 49
lon_min = -125
lon_max = -93

# Cleaning data

# Drop all rows where genus != Acamptopappus
print(f"shape pre genus filter: {acamptopappus_symbio_data_filtered.shape}")
acamptopappus_symbio_data_filtered = acamptopappus_symbio_data_filtered[
    acamptopappus_symbio_data_filtered["genus"] == "Acamptopappus"
]
print(f"shape post genus filter: {acamptopappus_symbio_data_filtered.shape}")

# Filter to western U.S.
print(f"shape pre lat/lon filter: {acamptopappus_symbio_data_filtered.shape}")
acamptopappus_symbio_data_filtered = acamptopappus_symbio_data_filtered[
    (acamptopappus_symbio_data_filtered["decimalLatitude"] >= lat_min) &
    (acamptopappus_symbio_data_filtered["decimalLatitude"] <= lat_max) &
    (acamptopappus_symbio_data_filtered["decimalLongitude"] >= lon_min) &
    (acamptopappus_symbio_data_filtered["decimalLongitude"] <= lon_max)
].copy()
print(f"shape post lat/lon filter: {acamptopappus_symbio_data_filtered.shape}")

# Filter locality if data is NaN, as this implies location data is likely poor
print(f"shape pre locality filter: {acamptopappus_symbio_data_filtered.shape}")
acamptopappus_symbio_data_filtered = acamptopappus_symbio_data_filtered[
    acamptopappus_symbio_data_filtered["locality"].notnull()
].copy()
print(f"shape post locality filter: {acamptopappus_symbio_data_filtered.shape}")

# Filter any entries after 1980
print(f"shape pre date filter: {acamptopappus_symbio_data_filtered.shape}")
acamptopappus_symbio_data_filtered = acamptopappus_symbio_data_filtered[
   acamptopappus_symbio_data_filtered["eventDate"] >= "1980-01-01"
].copy()
print(f"shape post date filter: {acamptopappus_symbio_data_filtered.shape}")

# Removing duplicates
print(f"shape pre duplicates filter: {acamptopappus_symbio_data_filtered.shape}")
acamptopappus_symbio_data_filtered = acamptopappus_symbio_data_filtered.drop_duplicates(
    subset = ["eventDate", "decimalLatitude", "decimalLongitude"],
    keep = "first"
)
print(f"shape post duplicates filter: {acamptopappus_symbio_data_filtered.shape}")
```

```python
# Plotting data
# Map the remaining records
fig = px.scatter_geo(
    acamptopappus_symbio_data_filtered,
    lat="decimalLatitude",
    lon="decimalLongitude",
    scope="usa",
    projection="albers usa",
    hover_name="genus",
    hover_data={
        "eventDate": True,
        "locality": True,
        "decimalLatitude": True,
        "decimalLongitude": True
    },
    title="Occurrences in the Western United States"
)

fig.update_traces(
    marker=dict(
        size=5,
        opacity=0.6
    )
)

fig.show()
```
